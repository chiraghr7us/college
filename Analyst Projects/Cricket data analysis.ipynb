```python
# importing libraries
import pandas as pd
import json
```

##  (1)Process Match Results


```python
with open('data-analytics-project-for-beginners/t20_json_files/t20_json_files/t20_wc_match_results.json') as f:
    data = json.load(f)

df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
# check no. of elements
df_match.shape
```

```python
# processing (treat score card as primary key and rename it)
df_match.rename({'scorecard': 'match_id'}, axis=1, inplace= True) # axis =1 means column is renamed
df_match.head(10)
```

## Create dictionary to store team1 vs team2 as key and match id as value

```python
match_ids_dict= {}

for index, row in df_match.iterrows():
    key1 =row['team1'] + ' Vs ' + row['team2']
    key2 =row['team2'] + ' Vs ' + row['team1'] # the way team is represented can be both ways

    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']

match_ids_dict
```

```python
df_match.to_csv('match_summary.csv', index = False)
```

## **(2) Process Batting summary**

```python
with open('data-analytics-project-for-beginners/t20_json_files/t20_json_files/t20_wc_batting_summary.json') as f:
    data = json.load(f)

    all_records = []
    for rec in data:
        all_records.extend(rec['battingSummary']) # adding arrays of all records
df_batting = pd.DataFrame(all_records)
df_batting.head(11)
```

```python
# covert dismilssal column into out/not out
df_batting["out/not_out"] = df_batting.dismissal.apply(lambda x : "out" if len(x) > 0 else "not out")
df_batting.head(10)
```

```python
# dropping dismissla column
df_batting.drop(columns = ['dismissal'] ,inplace =True) 
```

```python
df_batting.head(10)
```

```python
# removing unwanted characters from batsmanName
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x : x.replace('†' , ''))
```

```python
df_batting.head(11)
```

```python
df_batting['match_id']= df_batting['match'].map(match_ids_dict) # mapping match column to resopective match_id in dictionary
```

```python
df_batting.head(10)
```

```python
df_batting.to_csv('fact_batting_summary.csv', index= False)
```

## (3) Process Bowling summary

```python
with open('data-analytics-project-for-beginners/t20_json_files/t20_json_files/t20_wc_bowling_summary.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['bowlingSummary'])
all_records[:2]
```

## (4) process player info

```python
with open('data-analytics-project-for-beginners/t20_json_files/t20_json_files/t20_wc_player_info.json') as f:
    data = json.load(f)
```

```python
df_players = pd.DataFrame(data)
df_players.head()
```

```python
# cleaing up unwanted characters
df_players['name'] = df_players['name'].apply(lambda x: x.replace('â€', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('†', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('\xa0', ''))
df_players.head(10)
```
