```python
# ARRAY FUNCTIONS
a3= np.array([1,2,3])
print(a3)
a3=np.append(a3, [4,5,6]) #append- add to end of array
print(a3)
a3= np.insert(a3,4, [10,11,12]) # insert element in particular index
print(a3)
```

```python
a4=np.array([[1,2,3],
             [4,5,6]])
print(a4.ndim)
print(np.delete(a4, 0,0)) # delete 0th index row
print(np.delete(a4, 0 ,1)) # delete 0th index column
print(np.delete(a4, 1 ,1)) # delete 1st index column
```

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
#if using a Jupyter notebook, include:
%matplotlib inline
x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)
X,Y = np.meshgrid(x,y)
Z = X*np.exp(-X ,- Y)


fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d')


# Plot a 3D surface
ax.plot_surface(X, Y, Z)

```
